```python
rows, columns = trainset_X.shape
aux = np.zeros((rows, columns+1))
aux[:,:-1] = trainset_X
aux[:,-1] = trainset_Y
trainset = aux
trainset.shape
```

```python
rows, colunms = testset_X.shape
aux = np.zeros((rows, columns+1))
aux[:,:-1] = testset_X
aux[:,-1] = testset_Y
testset = aux
testset.shape
```

```python
#CÃ¡lculo da distÃ¢ncia euclidiana
def euclidian_dist(a, b):
    soma = 0
    for i in range(len(a)-1):
        soma += (a[i] - b[i])**2
    return soma **(1/2)
```

```python
#ImplementaÃ§Ã£o KNN
def getNeighbors(trainset, test, k):
    d = []
    neighbors = []
    for i in range(len(trainset)):
        #Calcula a distÃ¢ncia
        euclidian_d = euclidian_dist(test, trainset[i])
        #Lista com a instÃ¢ncia do conjunto de treino e sua distÃ¢ncia da instÃ¢ncia dada
        d.append((trainset[i], euclidian_d))
    #OrdenaÃ§Ã£o colocando os mais prÃ³ximos no comeÃ§o
    d.sort(key=operator.itemgetter(1))
    #Armazena as k primeiras instÃ¢ncias
    for j in range(k):
        neighbors.append(d[j][0])
    return neighbors
```

```python
#ClassificaÃ§Ã£o KNN
def classificKNN(neighbors):
    result = {}
    for i in range(len(neighbors)):
        #Pega o ultimo elemento de cada lista que Ã© a classe
        resp = neighbors[i][-1]
        if resp in result:
            result[resp] += 1
        else:
            result[resp] = 1
    resultMax = sorted(result.items(), key=operator.itemgetter(1), reverse=True)
    return resultMax[0][0]
```

```python
#Calcula a acurÃ¡cia do modelo
def getAccuracy(testset, predict):
    correct = 0
    for i in range(len(testset)):
        if predict[i] == testset[i][-1]:
            correct += 1
    return (correct/len(testset))*100
```

```python
k = 13
classification = []
for i in range(len(testset)):
    neighbors = getNeighbors(trainset, testset[i], k)
    result = classificKNN(neighbors)
    classification.append(result)
    print(f'Resultado retornado: {result.astype(int)} | Resultado esperado: {testset[i][-1].astype(int)}')
```

```python
accuracy = getAccuracy(testset, classification)
print(f'AcurÃ¡cia = {accuracy} %')
```
